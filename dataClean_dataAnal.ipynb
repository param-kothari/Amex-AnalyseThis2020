```python
dataTypeDict = dict(data.dtypes) #Dictionary containing column names as keys and data-types as values
```

```python
colObj_list = []#List containing "object" data-type column names
for var in dataTypeDict:
    if dataTypeDict[var] == 'object':
        data[var] = data[var].str.replace(',', '')
        data[var] = data[var].str.replace(' ', '')
        data[var] = data[var].str.replace('-', '0')
        data[var] = data[var].str.replace('(', '-')
        data[var] = data[var].str.replace(')', '')
        colObj_list.append(var)
        if (var != 'acq_sub_chn' and var!= 'acq_type_grp' and var!= 'fee_type_grp'):
            dataTypeDict[var] = 'float'
```

```python
data = data.astype(dataTypeDict) #Converting all numerical objects (numbers in the form of strings) 
                                #except 3 columns which contained alphabet-strings
#NOTE: DO NOT USE dataTypeDict anymore.
```

```python
data.isnull().sum(axis=0) #Sum of NaN values in each column of the main dataframe
```

```python
data = data.dropna() #Dataset after dropping all rows with atleast 1 NaN value
data.isnull().sum(axis=0) #Checking how many NaN values
```

```python
#Creating new column
data_new = data
cnsumr_chrg_avg = data['cnsumr_chrg_avg_credit_12m_amt']/data['cnsumr_chrg_actv_cust_cnt']
data_new['cnsumr_chrg_avg'] = cnsumr_chrg_avg
data_new['cnsumr_chrg_avg'] = data_new['cnsumr_chrg_avg'].fillna(0) #Replacing NaN values (dividing by zero) with zero
del data_new['cnsumr_chrg_avg_credit_12m_amt']
del data_new['cnsumr_chrg_actv_cust_cnt']
```

```python
cust_login = data['pre6m_cust_myca_active']+data['pre6m_cust_mob_logins']
data_new['cust_login'] = cust_login
del data_new['pre6m_cust_myca_active']
del data_new['pre6m_cust_mob_logins']
data_new.info()
```

```python
total_ext_cards = data['sow_tot_trans_cnt'] + data['sow_tot_revol_cnt']

rev_bal_per_ext_card = data['sow_revol_avg_paydown_pct']/total_ext_cards
Avg_pay_per_ext_card = data['sow_tot_annual_ext_pmt_amt']/total_ext_cards

data_new['rev_bal_per_ext_card'] = rev_bal_per_ext_card
data_new['Avg_pay_per_ext_card'] = Avg_pay_per_ext_card 

del data_new['sow_revol_avg_paydown_pct']
del data_new['sow_tot_annual_ext_pmt_amt']
del data_new['sow_tot_revol_cnt']
del data_new['sow_tot_trans_cnt']

data_new['rev_bal_per_ext_card'] = data_new['rev_bal_per_ext_card'].fillna(0)
data_new['Avg_pay_per_ext_card'] = data_new['Avg_pay_per_ext_card'].fillna(0)

data_new.info()
```

```python
data_new.isnull().sum(axis=0)
```

```python
total_spend_per_tran = data['pre6m_cust_spend']/data['pre6m_cust_roc_cnt']
data_new['total_spend_per_tran'] = total_spend_per_tran

data_new['total_spend_per_tran'] = data_new['total_spend_per_tran'].fillna(0)

del data_new['pre6m_cust_roc_cnt']
del data_new['pre6m_cust_spend']
```

```python
total_nondisc_per_tran = data['pre6m_cust_non_disc_amt']/data['pre6m_cust_non_disc_cnt']
data_new['total_nondisc_per_tran'] = total_nondisc_per_tran

data_new['total_nondisc_per_tran'] = data_new['total_nondisc_per_tran'].fillna(0)

total_disc_per_tran = data['pre6m_cust_disc_amt']/data['pre6m_cust_disc_cnt']
data_new['total_disc_per_tran'] = total_disc_per_tran

data_new['total_disc_per_tran'] = data_new['total_disc_per_tran'].fillna(0)

del data_new['pre6m_cust_non_disc_amt']
del data_new['pre6m_cust_non_disc_cnt']
del data_new['pre6m_cust_disc_amt']
del data_new['pre6m_cust_disc_cnt']
```
